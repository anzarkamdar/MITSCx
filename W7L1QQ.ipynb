```python
import numpy as np
import pandas as pd
import scipy.stats as stats
```

# QQ2 Empirical Distribution

```python
#df = pd.read_csv('QQ2.csv')
x = np.arange(20,29,1)
y = [0.05,0.15,0.22,0.17,0.12,0.09,0.09,0.08,0.03]
df = pd.DataFrame({"Lobster Demand":x,"Probability":y})
df['Cumsum']=df['Probability'].cumsum()
pSO = .05
pSOs=(df['Cumsum']>=(1-pSO)).idxmax()
q = 5
qs=1-df.loc[q,'Cumsum']
print(pSOs)
print(qs)

q = 15
df['Expected Sold'] = df.apply(lambda x: x['Probability']*x['Lobster Demand'] if x['Lobster Demand']<=q else x['Probability']*q,axis=1)
df['Expected Short'] = df.apply(lambda x: x['Probability']*0 if x['Lobster Demand']<=q else x['Probability']*(x['Lobster Demand']-q),axis=1)
Esold=df['Expected Sold'].sum()
Eshort=df['Expected Short'].sum()
Edemand = np.sum(df.apply(lambda x: x['Probability']*x['Lobster Demand'],axis=1))
print(Esold)
print(Eshort)
print(Edemand)
Esold+Eshort

df.head(9)
```

 # QQ4 Poisson Distribution

```python
lamda = 4

x = np.arange(20)
y = stats.poisson.pmf(x,lamda)
z = stats.poisson.cdf(x,lamda)
df3 = pd.DataFrame({'Chicken Demand':x,'Probability':y,'Cumulative Probability': z})

pSO = .05
q = 5
qs=(df3['Cumulative Probability']>=(1-pSO)).idxmax()
pSOs=1-df3.loc[q,'Cumulative Probability']
print('Number of chickens for P[SO] <= ' + str(pSO*100) + '% = ' + str(qs))
print('[PSO] if '+ str(q) + ' chickens are roasted = '+ str('%.2f'%(pSOs*100)) + '%')

df3['Expected Sold'] = df3.apply(lambda x: x['Probability']*x['Chicken Demand'] if x['Chicken Demand']<=q else x['Probability']*q,axis=1)
df3['Expected Short'] = df3.apply(lambda x: x['Probability']*0 if x['Chicken Demand']<=q else x['Probability']*(x['Chicken Demand']-q),axis=1)
Esold=df3['Expected Sold'].sum()
Eshort=df3['Expected Short'].sum()
Edemand = np.sum(df3.apply(lambda x: x['Probability']*x['Chicken Demand'],axis=1))
print('Esold = ' + str('%.2f' %Esold))
print('Eshort = ' + str('%.2f'%Eshort))
print('Edemand = ' + str('%.2f'%Edemand))
Esold+Eshort

df3.head()

```

# QQ5-QQ7 Continuous Uniform Distribution

```python
min = 80
max = 170
pSO = .05
q = 95

b=stats.uniform(min,max-min)
print("Mean Demand = " + str('%.1f'%b.mean()))
qs = b.ppf(1-pSO)
print('Number of chickens for P[SO] <= ' + str(pSO*100) + '% : ' + str(qs))
pSOs = 1-b.cdf(q)
print('[PSO] if '+ str(q) + ' chickens are roasted = '+ str('%.1f'%(pSOs*100)) + '%')

Esold = (q**2-min**2)/(2*(max-min)) + q*(max-q)/(max-min)
Eshort = (max-q)**2/(2*(max-min))

print('Esold = ' + str('%.2f' %Esold))
print('Eshort = ' + str('%.2f'%Eshort))

```

# QQ8 Normal Distribution

```python
mu = 22.55
sigma = 2.64

pSO = .05
q = 100

c = stats.norm(mu,sigma)

qs = c.ppf(1-pSO)
print('Number of chickens for P[SO] <= ' + str(pSO*100) + '% : ' + str('%.2f'%qs))
pSOs = 1-c.cdf(q)
print('[PSO] if '+ str(q) + ' chickens are roasted = '+ str('%.1f'%(pSOs*100)) + '%')

k = (q-mu)/sigma
Gk =stats.norm.pdf(k,0,1)- k*(1-stats.norm.cdf(k,0,1))
Eshort = sigma*Gk
Eshort
Edemand = c.mean()
Esold = Edemand-Eshort

print('Esold = ' + str('%.2f' %Esold))
print('Eshort = ' + str('%.2f'%Eshort))
print('Edemand = ' + str('%.2f'%Edemand))
```
